### To implement Polynomial Regression from scratch
1. Create a nx1 matrix for features.
2. Generate noise factor of nx1 dimension.
3. Generate three coefficients b0, b1, and b2 to define a quadratic equation.
4. Compute the polynomial regression model using the normal equation.
5. Plot the graph for visualization.

### Polynomial Regression [Scratch]

### Importing Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

### Generating Feature Matrix

```python
np.random.seed(42)
n = 100
X = np.random.rand(n, 1) * 10 
X[:10]
```

### Noise Vector

```python
noise = np.random.randn(n, 1) * 2
noise[:10]
```

### Defining Parameters

```python
b0 = 5  
b1 = 2  
b2 = 3  
y = b2 * X**2 + b1 * X + b0 + noise
y[:10]
```

### Scatter Plot

```python
plt.scatter(X, y, label="Original Data", color='blue')
plt.xlabel("X (Feature)")
plt.ylabel("Y (Target)")
plt.title("Scatter Plot of Data")
plt.legend()
plt.show()
```

### Computing Polynomial Features

```python
X_poly = np.hstack((np.ones((n, 1)), X, X**2))  
theta = np.linalg.inv(X_poly.T @ X_poly) @ X_poly.T @ y
```

### Prediction

```python
X_pred = np.linspace(0, 10, 100).reshape(-1, 1)
X_pred_poly = np.hstack((np.ones((100, 1)), X_pred, X_pred**2))
y_pred = X_pred_poly @ theta
y_pred[:10]
```

### Visualising Polynomial Regression

```python
plt.scatter(X, y, label="Original Data", color='blue')
plt.plot(X_pred, y_pred, label="Polynomial Regression (Degree 2)", color='red')
plt.xlabel("X (Feature)")
plt.ylabel("Y (Target)")
plt.title("Polynomial Regression Fit")
plt.legend()
plt.show()
```
