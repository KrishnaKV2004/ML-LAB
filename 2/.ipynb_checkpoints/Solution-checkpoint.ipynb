### Problem Statement

1. Create a dataset with one categorical feature[Gender] with some None values,
   one numerical feature[Age] with some Nan values.
2. Impute The Gender Column with the mode of the column and Age column with mean.
3. Visualize the data before and after imputation.

### Importing Libraries

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

### Generating Dataset

```python
gender_choices = ["Male", "Female", None]

data = {
    "Gender": [np.random.choice(gender_choices) for _ in range(100)],
    "Age": [np.random.choice([25, 30, 45, 50, 22, 27, 40, np.nan]) for _ in range(100)]
}

data = pd.DataFrame(data)
data
```

### Before Imputation

```python
plt.figure(figsize=(14, 6))
plt.subplot(1, 2, 1)
sns.countplot(data=data, x='Gender')
plt.title("Gender Distribution Before Imputation")

plt.subplot(1, 2, 2)
sns.histplot(data['Age'].dropna(), kde=True)
plt.title("Age Distribution Before Imputation")
plt.show()
```

### Imputation

```python
data['Gender'] = data['Gender'].fillna(data['Gender'].mode()[0])
data['Age'] = data['Age'].fillna(data['Age'].mean())
data
```

```python
print(data.isnull().sum())
```

### After Imputation

```python
plt.figure(figsize=(14, 6))
plt.subplot(1, 2, 1)
sns.countplot(data=data, x='Gender')
plt.title("Gender Distribution After Imputation")

plt.subplot(1, 2, 2)
sns.histplot(data['Age'], kde=True)
plt.title("Age Distribution After Imputation")
plt.show()
```
